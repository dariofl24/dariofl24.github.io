```python
def getRate(country):
    data = pd.read_csv("/Users/dflores/dariofl24/git/d3world/data_temps/"+country+".csv")
    
    data= data[data['date'] >= 1900 ]
    data= data.reset_index(drop=True)

    
    allDates=np.eye(len(data['date']),1)
    
    for idx in range(len(data['date'])):
        allDates[idx,0]=data['date'][idx]
        
    X=allDates
    y=data['value']
    
    reg = LinearRegression().fit(X, y)
    
    val_0 = reg.predict(np.array([[0]]))
    val_1 = reg.predict(np.array([[1]]))
    
    m = val_1 - val_0
    print(country+","+ str(m[0]*1000))
    
for code in codes:
  getRate(code)
```
